## Creating histograms
### Make a histogram from a pandas Series object

```python
address = '../mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg = cars.mpg

mpg.iplot(kind='histogram', filename='simple-histogram-chart')
```
