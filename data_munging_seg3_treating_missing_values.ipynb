# Data Munging Basics

## Segment 2 - Treating missing values

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

### Figuring out what data is missing

```python
missing = np.nan
series_obj = Series(['row1','row2',missing,'row4','row5','row6',missing,'row8'])
```

```python
series_obj
```

```python
series_obj.isnull()
```

### Filling in for missing values

```python
np.random.seed(25)
df_obj = DataFrame(np.random.randn(36).reshape(6,6))
df_obj
```

```python
df_obj.ix[3:5,0] = missing
df_obj.ix[1:4,5] = missing
df_obj
```

```python
filled_df = df_obj.fillna(0)
filled_df
```

```python
filled_df = df_obj.fillna({0:0.1,5: 1.25})
filled_df
```

```python
fill_df = df_obj.fillna(method='ffill')
fill_df
```

### Counting missing values

```python
np.random.seed(25)
df_obj = DataFrame(np.random.randn(36).reshape(6,6))
df_obj.iloc[3:5,0] = missing
df_obj.iloc[1:4,5] = missing

df_obj
```

```python
df_obj.isnull().sum()
```

### Filtering out missing values

```python
# value가 비어있지 않은 행만 리턴
df_no_nan = df_obj.dropna()
df_no_nan
```

```python
df_obj.dropna(how='all')
```
