### 데이터 추출 시 map함수를 이용해 특정컬럼을 SHA256 해시하기

```python
import pandas as pd
import numpy as np
import hashlib
```

```python
df = pd.read_csv('mtcars.csv',encoding='utf-8',dtype=object)
df.columns = columns=['cars','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
df
```

```python
### 해시 함수 추가
def hash_data(column):
    return hashlib.sha256(str.encode(column)).hexdigest()
```

```python
df['cars'] = df['cars'].map(hash_data)
df
```
